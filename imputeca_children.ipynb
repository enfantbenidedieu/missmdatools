```python
import pyreadr
result = pyreadr.read_r('./data/children.rda')
children = result["children"]
children
```

```python
for col in children.columns:
    children[col] = children[col].astype("float")
children.info()
```

```python
from missmdatools import imputeCA
res_impute = imputeCA(children,n_components=2,random_state=123)
```
